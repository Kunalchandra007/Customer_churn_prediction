```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\ML Byself\Deep Learning\customer_churn_dataset-testing-master.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
# Convert categorical columns to numerical
df['Gender'] = le.fit_transform(df['Gender'])
df['Subscription Type'] = le.fit_transform(df['Subscription Type'])
df['Contract Length'] = le.fit_transform(df['Contract Length'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
input_data = df.iloc[:, :-1]  
output_data = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit_transform(input_data)
```

```python
input_data = pd.DataFrame(ss.fit_transform(input_data), columns=input_data.columns)
```

```python
input_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
#to check input node
input_data.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
print(input_data.columns)  # Verify the number of features
```

```python
 import tensorflow
```

```python
input_data = input_data.astype("float32")
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann=Sequential()
```

```python
ann.add(Dense(10,input_dim=input_data.shape[1],activation="relu"))
ann.add(Dense(6,activation="relu"))
ann.add(Dense(2,activation="relu"))
ann.add(Dense(1,activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=1000,epochs=10)
```

```python
pr1=ann.predict(x_train)
pr_data1=[]
for i in pr1:
    if i[0] >0.5:
        pr_data1.append(1)
    else:
        pr_data1.append(0)
```

```python
pr=ann.predict(x_test)
```

```python
pr_data=[]
for i in pr:
    if i[0] >0.5:
        pr_data.append(1)
    else:
        pr_data.append(0)
```

```python
pr_data
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pr_data)#testing accuracy
```
